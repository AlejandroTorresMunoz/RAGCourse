# Introduction

In this module we will learn how to implement RAG with agents. **An agent is a LLM that can execute multiple tools.** For example, create and send emails.

There are two types of agents:

-   *ReAct Agent*: Can be any LLM.
-   *OpenAI Agent*: Only OpenAI LLM's.

**A tool is a method that can be called by the agent, and can have multiple parameters**.

The steps that usually follows an agent are these:  

1. Divide the question into multiple sub-questions, like *Chain-of-Thought*. 
2. Draw some sort of plan. 
3. Figure out the right set of tools to call.

However, there is a challenge when the user introduces a complex query that a unique RAG agent cannot solve. To handle with this problem, we can create **multi-document agents**. This structure allows that, instead of using for example the *top-k* similarity documents to answer a question when you have a **lot of different documents**, create an agent for each subset of documents and ask to each of them to return a summarization for his *top-k* documents. With this solution, if we have *n* agents, we will have *n* summaries that can be processed by the *multi-document agent* to solve the user's question.

For example, for the question : *Compare and constrast Uber and Lyft's revenue growth*, we have to do the following:

1. Write down the question into multiple sub-questions ovear each set of tools. In this case, we will need information about Uber and Lyft.
2. Run RAG pipeline for Uber and RAG pipeline for Lyft.
3. Combine the answers at the end.
